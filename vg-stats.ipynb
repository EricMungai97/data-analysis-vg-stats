```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')

df
```

# Company that is the most common video game publisher?

```python
most_common_publisher=df['Publisher'].value_counts().head(1)
most_common_publisher
```

# The most common platform?

```python
most_common_platform = df['Platform'].value_counts().head(1)
most_common_platform
```

# The most common genre

```python
most_common_genre = df['Genre'].value_counts().head(1)
most_common_genre
```

# Top 20 highest grossing games

```python
highest_grossing_games = df[['Name', 'Global_Sales']].sort_values('Global_Sales', ascending=False).head(20)
highest_grossing_games
```

# Median for North American Video Game Sales

```python
NA_video_game_sales_median=df['NA_Sales'].median()
NA_video_game_sales_median
```

# Secondary output showing ten games surrounding the median sales output

```python
with_median = df[df["NA_Sales"] == NA_video_game_sales_median]

with_median_count = len(with_median)
mid_of_median = with_median_count // 2

ten_median_na_sellers = with_median[mid_of_median - 5:mid_of_median + 5]
ten_median_na_seller_names = list(ten_median_na_sellers.Name)

ten_median_na_seller_names
```

# For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America

```python
max = df.NA_Sales.max()
mean = df.NA_Sales.mean()
standard_deviation = df.NA_Sales.std()
result = (max - mean) / standard_deviation
result
```

# How the Nintendo Wii average number of sales compare with all other platforms

```python


Wii_mean_sales = df[df['Platform'] == 'Wii']['Global_Sales'].mean()

print('Wii platform average sales in millions ==> '+ str(Wii_mean_sales))

None_Wii_mean_sales = df[df['Platform'] != 'Wii']['Global_Sales'].mean()

print('Other platform average sales in millions ==> ' + str(None_Wii_mean_sales))

sales_ratio_Wii_to_None_Wii = Wii_mean_sales/None_Wii_mean_sales

print('The Wii platform global sales is ' + str(round(sales_ratio_Wii_to_None_Wii,2)) + ' times higher than other platforms global sales.')

```

#  Based on Global Sales, who are the top 5 platforms

```python
df[['Platform','Global_Sales']].groupby('Platform').sum().sort_values('Global_Sales',ascending =False).head()
```
